```python
import pandas as pd
import numpy as np

X_titanic_test = pd.read_csv('../data/raw/test.csv')
y_titanic_test = pd.read_csv('../data/raw/gender_submission.csv')
X_titanic_train = pd.read_csv('../data/raw/train.csv')

#Removemos culumnas que en su mayoría son únicas de esta forma añadiendo ruido al modelo
X_titanic_test = X_titanic_test.drop('Name', axis=1)
X_titanic_test = X_titanic_test.drop('Ticket', axis=1)
y_titanic_test = y_titanic_test.drop('PassengerId', axis=1)

y_titanic_train = X_titanic_train['Survived']
X_titanic_train = X_titanic_train.drop('Survived', axis=1)
X_titanic_train = X_titanic_train.drop('Name', axis=1)
X_titanic_train = X_titanic_train.drop('Ticket', axis=1)



#Visualización de los datos para entender su estructura, en este caso vemos que Embarked y Sex son categóricos, Cabin tiene muchos valores distintos lo cual lo vuelve inutilizable por OHE
print("----------set of unique values in Defined column:")
words = set(" ".join(X_titanic_train['Embarked'].astype(str)).split())
print(len(words))
print(words)
words = set(" ".join(X_titanic_train['Cabin'].astype(str)).split())
print(len(words))
print(words)
words = set(" ".join(X_titanic_train['Sex'].astype(str)).split())
print(len(words))
print(words)

#Añadimos como visualización la información de los datos el porcentaje de NaN que tienen para entender bien los datos que faltan y que necesidades de "tratamiento" necesitan
print("----------info about X_titanic_train:", X_titanic_train.shape)
display(X_titanic_train.info())
display(X_titanic_train.head())
nan_percentages = X_titanic_train.isnull().sum() / len(X_titanic_train) * 100
print(nan_percentages)

print("----------info about y_titanic_train:", y_titanic_train.shape)
display(y_titanic_train.info())
display(y_titanic_train.head())

print("----------info about X_titanic_test:", X_titanic_test.shape)
display(X_titanic_test.info())
display(X_titanic_test.head())
nan_percentages = X_titanic_test.isnull().sum() / len(X_titanic_test) * 100
print(nan_percentages)

print("----------info about y_titanic_test:", y_titanic_test.shape)
display(y_titanic_test.info())
display(y_titanic_test.head())







                 
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder, StandardScaler
from sklearn.impute import KNNImputer

from sklearn.linear_model import LogisticRegression


scaler_pipeline = Pipeline([
    ('StandardScaler', StandardScaler())
])

preprocess = ColumnTransformer(
    transformers=[
        ('OneHotEncoder', OneHotEncoder(handle_unknown='ignore'), ['Embarked',"Sex"]),
        ('OrdinalEncoder', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1), ['Cabin']),
        ('StandardScaler',scaler_pipeline,['Age','Fare','SibSp','Parch']),
    ],
    remainder='passthrough' 
)

model = Pipeline([
    ('Preprocess', preprocess),
    ('Imputer', KNNImputer(n_neighbors=2)),
    ('Classifier', LogisticRegression(max_iter=2000))
])

print(y_titanic_test.shape)
model.fit(X_titanic_train, y_titanic_train)
print(model.score(X_titanic_test,y_titanic_test))
```

```python
from sklearn.model_selection import GridSearchCV


scoring = {
    'accuracy':'accuracy',
    'precision':'precision',
    'recall':'recall',
    'f1':'f1'
}

params = {
    'Classifier__C': [0.01, 0.1, 1, 10, 100],
    'Classifier__penalty': ['l2'],
    'Classifier__solver': ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga'],
    'Classifier__max_iter': [10,25,50,100,250,500,1000,1500,2000,5000],
    'Imputer__n_neighbors':[2,3,4,5,6,7]
}

gs = GridSearchCV(
    estimator=model,
    param_grid=params,
    cv=5,
    scoring=scoring,
    refit='f1',
    n_jobs=-1
)

gs.fit(X_titanic_train,y_titanic_train)

print("Best parameters:", gs.best_params_)
print("Best score:", gs.best_score_)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

best_model = gs.best_estimator_
best_model.fit(X_titanic_train, y_titanic_train)


scores = cross_val_score(model, X_titanic_train, y_titanic_train, cv=5)
print("Original model Cross-validation score",scores.mean())
scores = cross_val_score(best_model, X_titanic_train, y_titanic_train, cv=5)
print("GridSearchCV Cross-validation score",scores.mean())
print("\n")


y_pred_train = best_model.predict(X_titanic_train)
y_pred_test = best_model.predict(X_titanic_test)

print("Training Accuracy:",  accuracy_score(y_titanic_train, y_pred_train))
print("Testing Accuracy:",  accuracy_score(y_titanic_test, y_pred_test))
print("\n")
print("Training Precision:", precision_score(y_titanic_train, y_pred_train))
print("Testing Precision:", precision_score(y_titanic_test, y_pred_test))
print("\n")
print("Training Recall:",    recall_score(y_titanic_train, y_pred_train))
print("Testing Recall:",    recall_score(y_titanic_test, y_pred_test))
print("\n")
print("Training F1 Score:",  f1_score(y_titanic_train, y_pred_train))
print("Testing F1 Score:",  f1_score(y_titanic_test, y_pred_test))
```
