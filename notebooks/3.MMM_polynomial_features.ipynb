$$
y = f(x) =  \frac{1}{2} * X^2 + X + 2
$$

```python
import numpy as np
m = 100
X = 6 * np.random.rand(m,1)-3
y = 0.5 * X**2 + X + 2 + np.random.randn(m,1)
print(y.shape)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import SGDRegressor

for i in range(3): 
    poly = PolynomialFeatures(i+2,include_bias=True)
    poly_sgd = PolynomialFeatures(i+2,include_bias=False)
    X_poly = poly.fit_transform(X)
    X_poly_sgd = poly_sgd.fit_transform(X)



    reg = LinearRegression()
    sgdreg = SGDRegressor()
    reg.fit(X_poly,y)
    sgdreg.fit(X_poly_sgd,y)


    X_plot = np.linspace(-3,3,100).reshape(-1,1)
    X_plot_poly = poly.transform(X_plot)
    y_plot = reg.predict(X_plot_poly)

    X_plot_sgd = np.linspace(-3,3,100).reshape(-1,1)
    X_plot_sgd_poly = poly_sgd.transform(X_plot_sgd)
    y_plot_sgd = sgdreg.predict(X_plot_sgd_poly)


    plt.scatter(X,y)
    plt.plot(X_plot, y_plot, "b")
    plt.plot(X_plot_sgd, y_plot_sgd, "r")
    plt.show()
```
