```python
import kagglehub
import pandas as pd
```

```python
dataset_path = kagglehub.dataset_download("joebeachcapital/30000-spotify-songs")
spotify30k_df = pd.read_csv(dataset_path + "\\spotify_songs.csv")
```

```python
display(spotify30k_df.info())
display(spotify30k_df)
```

```python
#spotify30k_df["track_album_release_date"] = pd.to_datetime(spotify30k_df["track_album_release_date"],errors="coerce")


codes, uniques = pd.factorize(spotify30k_df['playlist_genre'])
spotify30k_df["playlist_genre_numeric"] = codes

codes, uniques = pd.factorize(spotify30k_df['playlist_subgenre'])
spotify30k_df["playlist_subgenre_numeric"] = codes

codes, uniques = pd.factorize(spotify30k_df['track_artist'])
spotify30k_df["track_artist_numeric"] = codes

spotify30k_df.dropna(inplace=True)


display(spotify30k_df.info())
display(spotify30k_df)
```

```python
import matplotlib.pyplot as plt

# track_id	
# track_name	
# track_artist	
# track_popularity	
# track_album_id	
# track_album_name	
# track_album_release_date	
# playlist_name	
# playlist_id	
# playlist_genre
# playlist:subgenre
# danceability
# energy
# key	
# loudness	
# mode	
# speechiness	
# acousticness	
# instrumentalness	
# liveness	
# valence	
# tempo	
# duration_ms
# track_artist_numeric
# playlist_subgenre_numeric
# playlist_genre_numeric

correlation_matrix = spotify30k_df.corr(numeric_only=True)

# Imprime la matriz de correlación
print(correlation_matrix.columns)

plt.matshow(correlation_matrix)
plt.show()

X = spotify30k_df[["loudness"]]
y = spotify30k_df[["energy"]]
plt.scatter(X,y,s=20)
```
