# Intro Gradiente Descendiente ITESO
$f(x) = x^2 + 1 $

$f'(x) = 2x$

```python
initial_x = 5
alpha = 0.1
n_iterations = 50

iterations = list()
y = list()

x = initial_x
for i in range(n_iterations):
    # calcular el gradiente para x actual
    g = 2*x
    #actualizar X con el gradiente actual
    x = x-alpha*g
    # guardar el valor de f(x)
    y.append(x**2 + 1)
    iterations.append(i+1)

    print(f"i: { i + 1}, x = {x}, y/f(x) = {x**2+1}")
    
```

```python
import matplotlib.pyplot as plt
from numpy import linspace

plt.subplot(1,2,1)
plt.plot(iterations, y)
plt.xlabel("Iterations/Epochs")
plt.ylabel("Loss/y/f(x)/J/C")

plt.subplot(1,2,2)
X = linspace(-5,5,100)
Y = X**2 + 1
plt.plot(X,Y,x,y[-1],"ro")
plt.xlabel("X")
plt.ylabel("Y")
```
